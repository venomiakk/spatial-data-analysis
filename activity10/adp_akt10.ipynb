<b>Ćwiczenie 1.</b> Utwórz interaktywną mapę na podstawie współrzędnych (60.25, 24.8).

```python
import folium

### your code ###
m = folium.Map(location=[60.25, 24.8])
m
```

```python
### your code ###
marker = folium.Marker(location=[60.20426, 24.96179], popup="Kumpula Campus")
marker.add_to(m)

m
```

<b>Ćwiczenie 4.</b> Zmień znacznik na zieloną ikonę OK.

<b>Ćwiczenie 5.</b> Wczytaj dane z pliku ``addresses.shp`` do obiektu typu GeoDataFrame o nazwie points.

```python
import geopandas as gpd

### your code ###
points = gpd.read_file('addresses/addresses.shp')
```

<b>Ćwiczenie 6.</b> Nanieś na interaktywną mapę o stylu cartodbpositron punkty ze zmiennej points. Przypisz im nazwę "Public transport stations".

```python
### your code ###
m = folium.Map(location=[60.25, 24.8], tiles='cartodbpositron')
for idx, row in points.iterrows():
    folium.Marker(
        location=[row.geometry.y, row.geometry.x],
        popup="Public transport stations"
    ).add_to(m)

m

# folium.GeoJson(points, popup="Public transport stations").add_to(m)
# m
```

<b>Ćwiczenie 7.</b> Dodaj do mapy możliwość kontroli warstw.

```python
### your code ###
folium.TileLayer("OpenStreetMap").add_to(m)
folium.LayerControl().add_to(m)
m
```

<b>Ćwiczenie 8.</b> Wczytaj do zmiennej locations pary współrzędnych (x, y) ze zmiennej points.

```python
### your code ###
locations = [(point.geometry.y, point.geometry.x) for point in points.itertuples()]
```

<b>Ćwiczenie 9.</b> Nanieś na interaktywną mapę o stylu cartodbpositron zgrupowane znaczniki o współrzędnych ze zmiennej locations.

```python
### your code ###
from folium.plugins import MarkerCluster
m = folium.Map(location=[60.25, 24.8], tiles='cartodbpositron')

marker_cluster = MarkerCluster().add_to(m)

for location in locations:
    folium.Marker(location=list(location), popup="Public transport stations").add_to(marker_cluster)

m
```
